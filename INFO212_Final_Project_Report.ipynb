### 6. Wrangling and Cleansing Data
---
*(Cleanse the data, fill up missing data, and wrangle the data to prepare for analysis. Explain the process and the results.)*

```python
# Check for missing values
missing_values = df.isnull().sum()
print("Missing Values:")
print(missing_values)

# Fill missing values in numerical columns with the mean
df['Age'].fillna(df['Age'].mean(), inplace=True)

# Fill missing values in categorical columns with the mode (most frequent value)
df['Flight Status'].fillna(df['Flight Status'].mode()[0], inplace=True)

# Verify that missing values have been filled
print("Missing Values After Filling:")
print(df.isnull().sum())
```

Explanation:

The isnull().sum() method is used to identify the columns with missing values
and display the count of missing values in each column.

For the 'Age' column (a numerical column), missing values are filled with the mean age.

For the 'Flight Status' column (a categorical column), missing values are filled with the most frequent value (mode).

```python
# Convert 'Departure Date' to datetime format
df['Departure Date'] = pd.to_datetime(df['Departure Date'], format='%m/%d/%Y')

# Verify the data types
print("Data Types:")
print(df.dtypes)
```

Explanation:

The 'Departure Date' column is converted to datetime format using pd.to_datetime().

This conversion allows for easier handling of date-related operations.

```python
# Remove duplicate rows
df.drop_duplicates(inplace=True)

# Verify the shape of the DataFrame after removing duplicates
print("Shape of DataFrame after removing duplicates:", df.shape)
```

Creating a new DataFrame df_c from the count_2 Series by converting the index and values into columns and renaming the columns.

pd.DataFrame(count_2): Converts the count_2 Series, which contains the counts of occurrences of each unique value in the "Country Name" column, into a DataFrame. The resulting DataFrame has two columns: the index (which is the unique country names) and a column named Country Name containing the counts.

Next, we rename the columns

```python
count_2 = df['Country Name'].value_counts()
```

```python
df_c = pd.DataFrame(count_2).reset_index()
```

```python
df_c = df_c.rename(columns={'index': 'Country Name', 'Country Name': 'count'})
```

```python
df_c.head()
```

Merging two DataFrames:

Merging two DataFrames, df_c and con, using the "Country Name" column. Let's break down the code:

con = df.loc[:, ['Country Name', 'Continents']]: Selects the "Country Name" and "Continents" columns from the original DataFrame df and creates a new DataFrame con.

con = con.drop_duplicates(): Removes duplicate rows based on the "Country Name" column in the con DataFrame. This ensures that each country appears only once.

new_df = pd.merge_ordered(df_c, con, fill_method='ffill', on='Country Name'): Performs an ordered merge on the "Country Name" column between the df_c DataFrame (containing counts) and the cleaned con DataFrame (containing unique country names and continents). The fill_method='ffill' parameter means forward fill, which fills missing values using the last known value.

```python
con=df.loc[:,['Country Name','Continents']]
con=con.drop_duplicates()
new_df=pd.merge_ordered(df_c,con, fill_method='ffil', on='Country Name')
```

```python
new_df.head()
```

Airport Country and Continent wise Passengers:

Using Plotly Express to create a sunburst chart to visualize the distribution of passengers across continents and countries. Let's break down the code:

Here's what each part of the code does:

px.sunburst(new_df, path=['Continents', 'Country Name'], values='count', ...): Creates a sunburst chart using Plotly Express. The path parameter specifies the hierarchy, with 'Continents' and 'Country Name' defining the levels. The values parameter represents the numerical values associated with each category, in this case, the count of passengers.

color='count': Colors the sunburst segments based on the count of passengers.

color_continuous_scale='RdBu': Sets the color scale for the continuous color mapping.

color_continuous_midpoint=np.average(new_df['count'], weights=new_df['count']): Specifies the midpoint of the color scale based on the average count, weighted by the count.

title="Airport Country and Continent wise Passengers": Sets the title of the sunburst chart.

fig.show(): Displays the generated sunburst chart.

```python
fig = px.sunburst(new_df, path=['Continents', 'Country Name'], values='count',
                 color='count',
                 color_continuous_scale='RdBu',
                 color_continuous_midpoint=np.average(new_df['count'], weights=new_df['count']),
                 title= "Airport Country and Continent wise Passengers")
fig.show()
```
